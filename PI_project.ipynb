```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Work', 'Family', 'Homeworks', 'Individual', 'Socializing', 'Spare Time', 'Sleep']
charles = [8.5, 0.5, 3.0, 1.0, 1.5, 3.0, 6.5]
henry = [9.5, 1.0, 2.0, 1.5, 0.5, 2.5, 7.0]
susan = [7.0, 1.5, 1.0, 2.5, 2.0, 2.0, 8.0]

# Plotting
x = np.arange(len(categories))
bar_width = 0.2

plt.bar(x, charles, width=bar_width, label='Charles')
plt.bar(x + bar_width, henry, width=bar_width, label='Henry')
plt.bar(x + 2 * bar_width, susan, width=bar_width, label='Susan')

plt.xlabel('Categories')
plt.ylabel('Hours')
plt.title('Area of Interest')
plt.xticks(x + bar_width, categories, rotation=45)
plt.legend()

# Set custom y-axis ticks
plt.yticks(np.arange(0, 10, 0.5))

plt.tight_layout()
plt.show()
```
